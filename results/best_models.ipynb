```python
import pandas as pd
models_df = pd.DataFrame({
    "Models": ["Majority Classifier", "KNN", "Decision Tree", "Random Forest", "Support Vector Machine", "XGBoost", "Tuned Random forest", "Tuned XGBoost", "Winning model"],
    "Private Accuracy": [0.81687, 0.92126, 0.89920, 0.93773, 0.93785, 0.93791, 0.93827, 0.93981, 0.94184]
})
models_df = models_df.sort_values("Private Accuracy", ascending=False)
```

```python
models_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=models_df, x="Private Accuracy", y="Models")
plt.title("Model Performance (Private Accuracy)")
plt.xlabel("Private Accuracy")
plt.ylabel("Models")
plt.xlim(0.8, 0.95)
plt.show()
```
