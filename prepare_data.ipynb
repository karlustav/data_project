```python
import pandas as pd
import numpy as np

data = pd.read_csv('train.csv')

data
```

```python
labels = data['Depression']
data.drop(columns=['Depression'], inplace=True)
data.drop(columns=['id'], inplace=True)
data.drop(columns=['Name'], inplace=True)
replace_dict = {
    'Gender': {'Male': 0, 'Female': 1},
    'Working Professional or Student': {'Student': 0, 'Working Professional': 1},
    'Academic Pressure': {'No': 0, 'Yes': 1},
    'Work Pressure': {'No': 0, 'Yes': 1},
    'Dietary Habits': {'Healthy': 0, 'Unhealthy': 1},
    'Family History of Mental Illness': {'No': 0, 'Yes': 1}
}

data = data.replace(replace_dict)
prepared_data = pd.concat([data, labels], axis=1)

data
```
