```python
import pandas as pd
import numpy as np

data = pd.read_csv('../data/train.csv')

data
```

# **Cleaning data**

### Gender mapping

```python
data['Gender'].value_counts()

data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
```

### Age mapping

```python
data.Age.value_counts()
#All good here
```

### City mapping

```python
# Using target encoding with cross-validation
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)

city_values = data['City'].copy()

for train_idx, val_idx in kf.split(data):
    city_encoding = data.iloc[train_idx].groupby('City')['Depression'].mean()
    city_values.iloc[val_idx] = data['City'].iloc[val_idx].map(city_encoding)

data['City'] = city_values
data['City'] = data['City'].fillna(data['Depression'].mean())
```

### Working professional or student mapping

```python
working_mapping = {
    'Student': 0,
    'Working Professional': 1
}
data['Working Professional or Student'] = data['Working Professional or Student'].map(working_mapping)
```

### Profession mapping

```python
#Mapping with target encoding

kf = KFold(n_splits=5, shuffle=True, random_state=42)

city_values = data['Profession'].copy()

for train_idx, val_idx in kf.split(data):
    city_encoding = data.iloc[train_idx].groupby('Profession')['Depression'].mean()
    city_values.iloc[val_idx] = data['Profession'].iloc[val_idx].map(city_encoding)

data['Profession'] = city_values
data['Profession'] = data['Profession'].fillna(data['Depression'].mean())
```

### Sleep duration mapping

```python

#Sleep duration mapping
sleep_mapping = {
    'Less than 5 hours': 4,
    '5-6 hours': 5.5,
    '7-8 hours': 7.5,
    'More than 8 hours': 9
}
data['Sleep Duration'] = data['Sleep Duration'].map(lambda x: x if x in sleep_mapping.keys() else np.nan)
data['Sleep Duration'] = data['Sleep Duration'].map(sleep_mapping)
data['Sleep Duration'] = data['Sleep Duration'].fillna(data['Sleep Duration'].median())
```

### Family history of mental illness mapping

```python
data["Family History of Mental Illness"].value_counts()
dict_family_history = {
    'No': 0,
    'Yes': 1
}
data["Family History of Mental Illness"] = data["Family History of Mental Illness"].map(dict_family_history)
# Korras
```

### Have you ever had suicidal thoughts mapping

```python
data["Have you ever had suicidal thoughts ?"].value_counts()
dict_suicidal_thoughts = {
    'No': 0,
    'Yes': 1
}
data["Have you ever had suicidal thoughts ?"] = data["Have you ever had suicidal thoughts ?"].map(dict_suicidal_thoughts)
```

### Degree mapping

```python
rare_degrees = data["Degree"].value_counts()[data["Degree"].value_counts() <= 20].index

print(len(data))
data = data[~data["Degree"].isin(rare_degrees)]


data["Degree"].value_counts()


kf = KFold(n_splits=5, shuffle=True, random_state=42)

data_encoded = data.copy()

data_encoded['Degree_encoded'] = np.nan

for train_idx, val_idx in kf.split(data):
    target_means = data.iloc[train_idx].groupby('Degree')['Depression'].mean()
    
    data_encoded.iloc[val_idx, data_encoded.columns.get_loc('Degree_encoded')] = \
        data.iloc[val_idx]['Degree'].map(target_means)

global_mean = data['Depression'].mean()
data_encoded['Degree_encoded'].fillna(global_mean, inplace=True)

print(data_encoded[['Degree', 'Degree_encoded', 'Depression']].head(10))

data_encoded.drop('Degree', axis=1, inplace=True)
data_encoded = data_encoded.rename(columns={'Degree_encoded': 'Degree'})

# print("\nVerifying the Degree column is now encoded:")
# print(data_encoded['Degree'].head())

data = data_encoded.copy()
data.head()
```

### Dietary habits mapping

```python
rare = data["Dietary Habits"].value_counts()[data["Dietary Habits"].value_counts() <= 20].index

# print(len(data))
data = data[~data["Dietary Habits"].isin(rare)]

# print("Shape before removing rare degrees:", len(data))
# print("\nDietary Habits removed:", list(rare_degrees))
data = pd.get_dummies(data, columns=['Dietary Habits'])
```

# Dropping columns

```python
# labels = data['Depression']
# data.drop(columns=['Depression'], inplace=True)
data.drop(columns=['id'], inplace=True)
data.drop(columns=['Name'], inplace=True)

```
