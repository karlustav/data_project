```python
import pandas as pd
import numpy as np

test_data = pd.read_csv('test.csv')
data = pd.read_csv('train.csv')

test_data
```

# **Cleaning data**

### Gender mapping

```python
test_data['Gender'].value_counts()

test_data['Gender'] = test_data['Gender'].map({'Male': 0, 'Female': 1})
```

### Age mapping

```python
test_data.Age.value_counts()
#All good here
```

### City mapping

```python
# Using target encoding with cross-validation
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Create a copy of the city values from the training data
city_values = test_data['City'].copy()

# Create a dictionary to hold the city encoding from the training data
city_encoding = {}

# Fit the KFold on the training data
for train_idx, val_idx in kf.split(data):
    # Calculate the mean depression value for each city in the training set
    city_encoding.update(data.iloc[train_idx].groupby('City')['Depression'].mean().to_dict())

# Map the city encoding to the test data
city_values = test_data['City'].map(city_encoding)

# Assign the encoded values back to the test data
test_data['City'] = city_values

# Fill NaN values with the mean of the depression values from the training data
test_data['City'] = test_data['City'].fillna(data['Depression'].mean())
```

### Working professional or student mapping

```python
working_mapping = {
    'Student': 0,
    'Working Professional': 1
}
test_data['Working Professional or Student'] = test_data['Working Professional or Student'].map(working_mapping)
```

### Profession mapping

```python
# Using target encoding with cross-validation
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Create a copy of the city values from the training data
city_values = test_data['Profession'].copy()

# Create a dictionary to hold the city encoding from the training data
city_encoding = {}

# Fit the KFold on the training data
for train_idx, val_idx in kf.split(data):
    # Calculate the mean depression value for each city in the training set
    city_encoding.update(data.iloc[train_idx].groupby('Profession')['Depression'].mean().to_dict())

# Map the city encoding to the test data
city_values = test_data['Profession'].map(city_encoding)

# Assign the encoded values back to the test data
test_data['Profession'] = city_values

# Fill NaN values with the mean of the depression values from the training data
test_data['Profession'] = test_data['Profession'].fillna(data['Depression'].mean())
```

### Sleep duration mapping

```python

#Sleep duration mapping
sleep_mapping = {
    'Less than 5 hours': 4,
    '5-6 hours': 5.5,
    '7-8 hours': 7.5,
    'More than 8 hours': 9
}
test_data['Sleep Duration'] = test_data['Sleep Duration'].map(lambda x: x if x in sleep_mapping.keys() else np.nan)
test_data['Sleep Duration'] = test_data['Sleep Duration'].map(sleep_mapping)
test_data['Sleep Duration'] = test_data['Sleep Duration'].fillna(test_data['Sleep Duration'].median())
```

### Degree mapping

```python
# First calculate target means from training data
target_means = data["Degree"].copy()

# Calculate global mean from training data
global_mean = data['Depression'].mean()  # Add this line

# Apply encoding to test data
test_data_encoded = test_data.copy()
rare_degrees = test_data["Degree"].value_counts()[test_data["Degree"].value_counts() <= 20].index
# Remove rare degrees from test data (same as training)
test_data = test_data[~test_data["Degree"].isin(rare_degrees)]

# Map the degrees using training data means
test_data_encoded['Degree_encoded'] = test_data['Degree'].map(target_means)

# Fill missing values with global mean from training data
test_data_encoded['Degree_encoded'].fillna(global_mean, inplace=True)

# Clean up columns
test_data_encoded.drop('Degree', axis=1, inplace=True)
test_data_encoded = test_data_encoded.rename(columns={'Degree_encoded': 'Degree'})

test_data = test_data_encoded.copy()
```

### Dietary habits mapping

```python
rare = test_data["Dietary Habits"].value_counts()[test_data["Dietary Habits"].value_counts() <= 20].index

# print(len(data))
test_data = test_data[~test_data["Dietary Habits"].isin(rare)]

# print("Shape before removing rare degrees:", len(data))
# print("\nDietary Habits removed:", list(rare_degrees))
test_data = pd.get_dummies(test_data, columns=['Dietary Habits'])
```
