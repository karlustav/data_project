```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('../data/train.csv')

data.info()

```

```python
plt.figure(figsize=(10, 5))
plt.hist(data['Gender'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist([data[data['Gender']=='Male']['Age'], data[data['Gender']=='Female']['Age']], label=['Male', 'Female'])
plt.legend()

plt.show()
```
